```python
#demand h supply nhi
'''
Stock Splits and Data Adjustment Practices
Stock splits increase the number of shares by dividing existing ones (e.g., a 1:4 split makes 1 share become 4). This results in a proportional decrease in the stock price and an increase in trading volume.
However, most data providers adjust historical prices to reflect splits, but do not always adjust historical volumes. This causes older volume figures to appear inflated, and more recent data to look lower by comparison — even though actual trading activity may not have dropped.
This inconsistency in volume adjustment is one of the primary reasons behind the perceived decline in volume.

Market Maturity and Institutional Holding
In the early growth phase of a company, trading activity tends to be higher due to retail speculation and market volatility. Over time, as the company matures, its stock becomes a long-term asset held by institutional investors.
These investors trade less frequently, leading to a natural decline in daily trading volume, even though price continues to rise. Hence, lower volume in recent years often reflects market stability rather than disinterest.

Shift in Investor Behavior
The trend toward buy-and-hold strategies, the growth of passive investing (e.g., ETFs), and algorithmic trading have changed volume patterns.
Institutional dominance and reduced speculative trades lead to smoother volume curves in mature stocks.
'''
```

```python
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt

# --- User Input for Company ---
company_input = input("Enter the company (AAPL, AMZN, GOOGL, MSFT, NVDA): ").upper()

# --- Column Mapping ---
column_map = {
    'AAPL': ['Date', 'Close_AAPL', 'High_AAPL', 'Low_AAPL', 'Open_AAPL', 'Volume_AAPL'],
    'AMZN': ['Date', 'Close_AMZN', 'High_AMZN', 'Low_AMZN', 'Open_AMZN', 'Volume_AMZN'],
    'GOOGL': ['Date', 'Close_GOOGL', 'High_GOOGL', 'Low_GOOGL', 'Open_GOOGL', 'Volume_GOOGL'],
    'MSFT': ['Date', 'Close_MSFT', 'High_MSFT', 'Low_MSFT', 'Open_MSFT', 'Volume_MSFT'],
    'NVDA': ['Date', 'Close_NVDA', 'High_NVDA', 'Low_NVDA', 'Open_NVDA', 'Volume_NVDA']
}

df_selected = df[column_map[company_input]].copy()
df_selected.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
df_selected['Date'] = pd.to_datetime(df_selected['Date'])
df_selected.set_index('Date', inplace=True)
df_selected.sort_index(inplace=True)

# --- Function to Detect and Plot Doji ---
def detect_and_plot_doji(data_df, title_suffix=""):
    temp_df = data_df.copy()

    temp_df['Range'] = temp_df['High'] - temp_df['Low']
    temp_df['Body'] = abs(temp_df['Open'] - temp_df['Close'])
    temp_df['MidBody'] = (temp_df['Open'] + temp_df['Close']) / 2
    temp_df['MidRange'] = (temp_df['High'] + temp_df['Low']) / 2
    temp_df['CenterDiff'] = abs(temp_df['MidBody'] - temp_df['MidRange'])

    # Tighter conditions for accurate Doji
    doji_body_ratio = 0.02  # Body must be <= 2% of range
    center_ratio = 0.1      # Body center must be near candle center

    temp_df['is_doji'] = (
        (temp_df['Range'] > 0) &
        (temp_df['Body'] <= doji_body_ratio * temp_df['Range']) &
        (temp_df['CenterDiff'] <= center_ratio * temp_df['Range'])
    )

    avg_range = temp_df['Range'].mean() if temp_df['Range'].mean() > 0 else 1
    offset = avg_range * 0.5

    doji_y = pd.Series(index=temp_df.index, dtype=float)
    doji_y[temp_df['is_doji']] = temp_df.loc[temp_df['is_doji'], 'Low'] - offset

    apds = []
    if not doji_y.dropna().empty:
        apds.append(mpf.make_addplot(doji_y, type='scatter', marker='^', color='blue', markersize=90, panel=0))
        print(f"✅ Detected {len(doji_y.dropna())} Doji pattern(s).")
    else:
        print("❌ No Doji patterns detected.")

    print(f"\n📊 Plotting {company_input} Candlestick Chart{title_suffix} with Doji Patterns...\n")
    mpf.plot(temp_df,
             type='candle',
             style='yahoo',
             volume=True,
             title=f"{company_input} Candlestick Chart{title_suffix} with Doji Patterns",
             addplot=apds,
             figscale=1.5)

# --- Resample to Monthly and Run ---
print(f"📅 Detecting Doji patterns on Monthly Chart for {company_input}...")

monthly_df = df_selected.resample('ME').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume': 'sum'
}).dropna()

# Call detection
detect_and_plot_doji(monthly_df, " - Monthly Chart")
print("✅ Done.")
```

```python
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt

# --- User Input for Company ---
company_input = input("Enter the company (AAPL, AMZN, GOOGL, MSFT, NVDA): ").upper()

# --- Column Mapping ---
column_map = {
    'AAPL': ['Date', 'Close_AAPL', 'High_AAPL', 'Low_AAPL', 'Open_AAPL', 'Volume_AAPL'],
    'AMZN': ['Date', 'Close_AMZN', 'High_AMZN', 'Low_AMZN', 'Open_AMZN', 'Volume_AMZN'],
    'GOOGL': ['Date', 'Close_GOOGL', 'High_GOOGL', 'Low_GOOGL', 'Open_GOOGL', 'Volume_GOOGL'],
    'MSFT': ['Date', 'Close_MSFT', 'High_MSFT', 'Low_MSFT', 'Open_MSFT', 'Volume_MSFT'],
    'NVDA': ['Date', 'Close_NVDA', 'High_NVDA', 'Low_NVDA', 'Open_NVDA', 'Volume_NVDA']
}

df_selected = df[column_map[company_input]].copy()
df_selected.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
df_selected['Date'] = pd.to_datetime(df_selected['Date'])
df_selected.set_index('Date', inplace=True)
df_selected.sort_index(inplace=True)

# --- Function to Detect and Plot Spinning Tops ---
def detect_and_plot_spinning_tops(data_df, title_suffix=""):
    temp_df = data_df.copy()

    # Calculate body and shadows
    temp_df['Body'] = abs(temp_df['Close'] - temp_df['Open'])
    temp_df['Range'] = temp_df['High'] - temp_df['Low']
    temp_df['UpperShadow'] = temp_df['High'] - temp_df[['Open', 'Close']].max(axis=1)
    temp_df['LowerShadow'] = temp_df[['Open', 'Close']].min(axis=1) - temp_df['Low']

    # Spinning Top Conditions
    body_threshold = 0.3
    shadow_threshold = 0.3
    temp_df['is_spinning_top'] = (
        (temp_df['Range'] > 0) &
        (temp_df['Body'] <= body_threshold * temp_df['Range']) &
        (temp_df['UpperShadow'] >= shadow_threshold * temp_df['Range']) &
        (temp_df['LowerShadow'] >= shadow_threshold * temp_df['Range'])
    )

    # Offset below the low
    avg_range = temp_df['Range'].mean() if temp_df['Range'].mean() > 0 else 1
    offset = avg_range * 0.5

    # Aligned y-values for plotting
    spin_y = pd.Series(index=temp_df.index, dtype=float)
    spin_y[temp_df['is_spinning_top']] = temp_df.loc[temp_df['is_spinning_top'], 'Low'] - offset

    apds = []
    if not spin_y.dropna().empty:
        apds.append(
            mpf.make_addplot(spin_y, type='scatter', marker='^', color='purple', markersize=90, panel=0)
        )
        print(f"✅ Detected {len(spin_y.dropna())} Spinning Top pattern(s).")
    else:
        print("❌ No Spinning Top patterns found.")

    # Plot the chart
    print(f"\n📊 Plotting Spinning Top Candlestick Chart{title_suffix} for {company_input}...\n")
    mpf.plot(temp_df,
             type='candle',
             style='yahoo',
             volume=True,
             title=f"{company_input} Candlestick Chart{title_suffix} with Spinning Top Patterns",
             addplot=apds,
             figscale=1.5)

# --- Resample to Monthly and Run ---
print(f"🟣 Generating Monthly Chart for {company_input} with Spinning Top patterns...")

monthly_df = df_selected.resample('ME').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume': 'sum'
}).dropna()

# Run the function
detect_and_plot_spinning_tops(monthly_df, " - Monthly Chart")

print("✅ Done.")
```

```python
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt

# --- User Input for Company ---
company_input = input("Enter the company (AAPL, AMZN, GOOGL, MSFT, NVDA): ").upper()

# --- Column Mapping ---
column_map = {
    'AAPL': ['Date', 'Close_AAPL', 'High_AAPL', 'Low_AAPL', 'Open_AAPL', 'Volume_AAPL'],
    'AMZN': ['Date', 'Close_AMZN', 'High_AMZN', 'Low_AMZN', 'Open_AMZN', 'Volume_AMZN'],
    'GOOGL': ['Date', 'Close_GOOGL', 'High_GOOGL', 'Low_GOOGL', 'Open_GOOGL', 'Volume_GOOGL'],
    'MSFT': ['Date', 'Close_MSFT', 'High_MSFT', 'Low_MSFT', 'Open_MSFT', 'Volume_MSFT'],
    'NVDA': ['Date', 'Close_NVDA', 'High_NVDA', 'Low_NVDA', 'Open_NVDA', 'Volume_NVDA']
}


df_selected = df[column_map[company_input]].copy()
df_selected.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
df_selected['Date'] = pd.to_datetime(df_selected['Date'])
df_selected.set_index('Date', inplace=True)
df_selected.sort_index(inplace=True)

# --- Function to Detect and Plot Paper Umbrella Patterns ---
def detect_and_plot_paper_umbrella(data_df, title_suffix=""):
    temp_df = data_df.copy()

    # Calculations
    temp_df['Range'] = temp_df['High'] - temp_df['Low']
    temp_df['Body'] = abs(temp_df['Close'] - temp_df['Open'])
    temp_df['UpperShadow'] = temp_df['High'] - temp_df[['Open', 'Close']].max(axis=1)
    temp_df['LowerShadow'] = temp_df[['Open', 'Close']].min(axis=1) - temp_df['Low']

    # Paper Umbrella Conditions
    body_threshold = 0.3
    lower_shadow_threshold = 2.0
    upper_shadow_max_ratio = 0.1

    temp_df['is_paper_umbrella'] = (
        (temp_df['Body'] <= body_threshold * temp_df['Range']) &
        (temp_df['LowerShadow'] >= lower_shadow_threshold * temp_df['Body']) &
        (temp_df['UpperShadow'] <= upper_shadow_max_ratio * temp_df['Range'])
    )

    temp_df['Hammer'] = temp_df['is_paper_umbrella'] & (temp_df['Close'] > temp_df['Open'])
    temp_df['HangingMan'] = temp_df['is_paper_umbrella'] & (temp_df['Open'] > temp_df['Close'])

    avg_range = temp_df['Range'].mean() if temp_df['Range'].mean() > 0 else 1
    offset = avg_range * 0.5

    hammer_y = pd.Series(index=temp_df.index, dtype=float)
    hanging_y = pd.Series(index=temp_df.index, dtype=float)

    hammer_y[temp_df['Hammer']] = temp_df['Low'] - offset
    hanging_y[temp_df['HangingMan']] = temp_df['Low'] - offset

    apds = []

    if not hammer_y.dropna().empty:
        apds.append(mpf.make_addplot(hammer_y, type='scatter', marker='^', color='green', markersize=90, panel=0))
        print(f"✅ Detected {len(hammer_y.dropna())} Hammer(s).")
    else:
        print("❌ No Hammer patterns found.")

    if not hanging_y.dropna().empty:
        apds.append(mpf.make_addplot(hanging_y, type='scatter', marker='v', color='red', markersize=90, panel=0))
        print(f"✅ Detected {len(hanging_y.dropna())} Hanging Man(s).")
    else:
        print("❌ No Hanging Man patterns found.")

    print(f"\n📊 Plotting {company_input} Candlestick Chart{title_suffix} with Paper Umbrella Patterns...\n")
    mpf.plot(temp_df,
             type='candle',
             style='yahoo',
             volume=True,
             title=f"{company_input} Candlestick Chart{title_suffix} with Hammer and Hanging Man",
             addplot=apds,
             figscale=1.5)

# --- Resample to Monthly and Run ---
print(f"📌 Detecting Paper Umbrella (Hammer + Hanging Man) patterns for {company_input}...")

monthly_df = df_selected.resample('ME').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume': 'sum'
}).dropna()

# Run Detection
detect_and_plot_paper_umbrella(monthly_df, " - Monthly Chart")
print("✅ Done.")
```

```python
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt

# --- User Input for Company ---
company_input = input("Enter the company (AAPL, AMZN, GOOGL, MSFT, NVDA): ").upper()

# --- Column Mapping ---
column_map = {
    'AAPL': ['Date', 'Close_AAPL', 'High_AAPL', 'Low_AAPL', 'Open_AAPL', 'Volume_AAPL'],
    'AMZN': ['Date', 'Close_AMZN', 'High_AMZN', 'Low_AMZN', 'Open_AMZN', 'Volume_AMZN'],
    'GOOGL': ['Date', 'Close_GOOGL', 'High_GOOGL', 'Low_GOOGL', 'Open_GOOGL', 'Volume_GOOGL'],
    'MSFT': ['Date', 'Close_MSFT', 'High_MSFT', 'Low_MSFT', 'Open_MSFT', 'Volume_MSFT'],
    'NVDA': ['Date', 'Close_NVDA', 'High_NVDA', 'Low_NVDA', 'Open_NVDA', 'Volume_NVDA']
}


df_selected = df[column_map[company_input]].copy()
df_selected.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
df_selected['Date'] = pd.to_datetime(df_selected['Date'])
df_selected.set_index('Date', inplace=True)
df_selected.sort_index(inplace=True)

# --- Function to Detect and Plot Shooting Stars ---
def detect_and_plot_shooting_stars(data_df, title_suffix=""):
    temp_df = data_df.copy()

    # Calculate shadows and body
    temp_df['Body'] = abs(temp_df['Close'] - temp_df['Open'])
    temp_df['Range'] = temp_df['High'] - temp_df['Low']
    temp_df['UpperShadow'] = temp_df['High'] - temp_df[['Open', 'Close']].max(axis=1)
    temp_df['LowerShadow'] = temp_df[['Open', 'Close']].min(axis=1) - temp_df['Low']

    # Thresholds
    body_ratio = 0.3
    upper_shadow_min = 2.0  # ≥ 2x body
    lower_shadow_max_ratio = 0.1  # ≤ 10% of range

    # Shooting Star condition
    temp_df['is_shooting_star'] = (
        (temp_df['Body'] <= body_ratio * temp_df['Range']) &
        (temp_df['UpperShadow'] >= upper_shadow_min * temp_df['Body']) &
        (temp_df['LowerShadow'] <= lower_shadow_max_ratio * temp_df['Range']) &
        (temp_df['Open'] > temp_df['Close'])  # bearish candle
    )

    # Offset to move marker above the candle
    avg_range = temp_df['Range'].mean() if temp_df['Range'].mean() > 0 else 1
    offset = avg_range * 0.5

    star_y = pd.Series(index=temp_df.index, dtype=float)
    star_y[temp_df['is_shooting_star']] = temp_df.loc[temp_df['is_shooting_star'], 'High'] + offset

    apds = []
    if not star_y.dropna().empty:
        apds.append(mpf.make_addplot(star_y, type='scatter', marker='v', color='red', markersize=90, panel=0))
        print(f"✅ Detected {len(star_y.dropna())} Shooting Star(s).")
    else:
        print("❌ No Shooting Star patterns detected.")

    # Plot chart
    print(f"\n📊 Plotting Shooting Star Candlestick Chart{title_suffix} for {company_input}...\n")
    mpf.plot(temp_df,
             type='candle',
             style='yahoo',
             volume=True,
             title=f"{company_input} Candlestick Chart{title_suffix} with Shooting Star Patterns",
             addplot=apds,
             figscale=1.5)

# --- Resample to Monthly and Run ---
print(f"🌟 Detecting Shooting Star patterns on Monthly Chart for {company_input}...")

monthly_df = df_selected.resample('ME').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume': 'sum'
}).dropna()

detect_and_plot_shooting_stars(monthly_df, " - Monthly Chart")
print("✅ Done.")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Prepare Data
# -------------------------------
df1 = df[['Date', 'Close_GOOGL', 'High_GOOGL','Low_GOOGL', 'Open_GOOGL',
          'Volume_GOOGL','Return_GOOGL','Open_Close_Change_GOOGL', 'Daily_Range_GOOGL']].copy()
df1 = df1.sort_values('Date').reset_index(drop=True)

df1['Lag1'] = df1['Close_GOOGL'].shift(1)
df1['Lag2'] = df1['Close_GOOGL'].shift(2)

# 🔑 Correct Target
df1['Target'] = df1['Close_GOOGL'].shift(-1)

# Drop NaN rows
df1.dropna(inplace=True)

X = df1[['Lag1', 'Lag2', 'High_GOOGL','Low_GOOGL', 'Open_GOOGL',
          'Volume_GOOGL','Return_GOOGL','Open_Close_Change_GOOGL', 'Daily_Range_GOOGL']]
y = df1['Target']

# -------------------------------
# Train-Test Split
# -------------------------------
split_ratio = 0.8
split_index = int(len(df1) * split_ratio)

X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

# -------------------------------
# Model
# -------------------------------
model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=3,
    random_state=42
)
model.fit(X_train, y_train)

# -------------------------------
# Predict
# -------------------------------
y_pred = model.predict(X_test)

# -------------------------------
# Evaluate
# -------------------------------
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
accuracy = 100 - mape

print(f"✅ RMSE: {rmse:.2f}")
print(f"✅ R2 Score: {r2:.4f}")
print(f"✅ MAPE: {mape:.2f}%")
print(f"✅ Approx Accuracy: {accuracy:.2f}%")

# -------------------------------
# Plot
# -------------------------------
plt.figure(figsize=(14,6))
plt.plot(df1['Date'].iloc[split_index:], y_test, label="Actual")
plt.plot(df1['Date'].iloc[split_index:], y_pred, label="Predicted")
plt.title("AAPL Closing Price: Actual vs Predicted (Random Forest)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
